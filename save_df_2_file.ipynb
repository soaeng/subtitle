```python
import dotenv
import os
import pymongo
import pandas as pd
import numpy as np
```

```python
# í™˜ê²½ë³€ìˆ˜ ë¶ˆëŸ¬ì˜¤ê¸°
dotenv.load_dotenv(dotenv.find_dotenv())
USER = os.environ["MONGO_USER"] # MongoDB user
PASSWORD = os.environ["MONGO_PASS"] # MongoDB password
PORT = int(os.environ["MONGO_PORT"]) # MongoDB port

# DB ì—°ê²°
# conn = pymongo.MongoClient("localhost", 27017)
conn = pymongo.MongoClient("mongodb://" + USER + ":" + PASSWORD + "@j7e104.p.ssafy.io", PORT)
db = conn.animation
dbcol_info = db.ani_info

df = pd.DataFrame(dbcol_info.find({}))
df = df[["id", "air_year_quarter", "author", "avg_rating", "content", "content_rating", "distributed_air_time", "genres", "highlight_video", "images", "is_adult", "is_ending", "name", "production", "series_id", "tags"]]
df["series_id"] = pd.to_numeric(df["series_id"], errors="coerce")
df["series_id"] = df["series_id"].astype("float").astype("Int32")
df
```
